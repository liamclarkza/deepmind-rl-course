# Lecture 1: Introduction to Reinforcement Learning

## What is reinforcement learning?



Reinforcement Learning: Science and framework of **learning to make decisions** from **interaction**.

- At each timestep $t$, an agent:
  - Receives observation $O_t$
  - Executes action $A_t$
- The environment:
  - Receives action $A_t$
  - Emits observation $O_{t+1}$ (and reward $R_{t+1}$)
